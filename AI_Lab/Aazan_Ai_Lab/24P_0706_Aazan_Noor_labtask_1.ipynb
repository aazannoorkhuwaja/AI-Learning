```python
import matplotlib.pyplot as plt
plt.title("Straight Line and Cosine")
x=np.arange(0,11)
t=np.cos(x)
y=2*x+1
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y,marker='o')
plt.plot(x,t,marker='x')
#plt.grid()
plt.show()
```
